```python
class GradCAM:
    def __init__(self, model, target_layer):
        self.model = model
        self.target_layer = target_layer
        self.gradients = None
        self.activations = None
        
        # Register hooks
        target_layer.register_forward_hook(self.save_activations)
        target_layer.register_full_backward_hook(self.save_gradients)

    def save_activations(self, module, input, output):
        self.activations = output.detach()

    def save_gradients(self, module, grad_input, grad_output):
        self.gradients = grad_output[0].detach()

    def generate(self, input_image):
        # Forward pass
        output = self.model(input_image)
        pred_class = output.argmax(dim=1).item()
        
        # Backward pass for gradients
        self.model.zero_grad()
        one_hot = torch.zeros_like(output)
        one_hot[0][pred_class] = 1
        output.backward(gradient=one_hot)
        
        # Compute weights
        pooled_gradients = torch.mean(self.gradients, dim=[0, 2, 3])
        heatmap = torch.matmul(self.activations[0], pooled_gradients)
        heatmap = torch.relu(heatmap).squeeze().cpu().numpy()
        
        # Normalize and resize
        heatmap = cv2.resize(heatmap, (input_image.shape[3], input_image.shape[2]))
        heatmap = (heatmap - heatmap.min()) / (heatmap.max() - heatmap.min() + 1e-8)
        return heatmap, pred_class

# Initialize Grad-CAM (target last conv layer)
target_layer = model.features[-3]  # Adjust index to your model's last conv layer
gradcam = GradCAM(model, target_layer)
```

```python
# Preprocessing
preprocess = transforms.Compose([
        transforms.Resize(256),
        transforms.RandomRotation(20),
        transforms.RandomHorizontalFlip(),
        transforms.RandomVerticalFlip(),
        transforms.ColorJitter(brightness=0.2, contrast=0.2, saturation=0.2),
        transforms.RandomResizedCrop(224, scale=(0.8, 1.0)),
        transforms.ToTensor(),
        transforms.Normalize([0.485, 0.456, 0.406], [0.229, 0.224, 0.225])
    ])

# Perturbation functions
def add_gaussian_noise(image_tensor, intensity=0.1):
    noise = torch.randn_like(image_tensor) * intensity
    return torch.clamp(image_tensor + noise, 0, 1)

def add_gaussian_blur(image_tensor, kernel_size=5):
    image_np = image_tensor.permute(1, 2, 0).numpy()
    blurred = cv2.GaussianBlur(image_np, (kernel_size, kernel_size), 0)
    return torch.tensor(blurred).permute(2, 0, 1)

def add_occlusion(image_tensor, occ_size=50):
    occluded = image_tensor.clone()
    h, w = occluded.shape[1], occluded.shape[2]
    x = np.random.randint(0, w - occ_size)
    y = np.random.randint(0, h - occ_size)
    occluded[:, y:y+occ_size, x:x+occ_size] = 0
    return occluded
```

```python
# 3. Analysis pipeline
def analyze_model_behavior(image_path, model, preprocess):
    # Load and preprocess image
    orig_image = cv2.imread(image_path)
    image = preprocess(orig_image).unsqueeze(0)
    
    # Original prediction
    with torch.no_grad():
        outputs = model(image)
    pred_class = outputs.argmax().item()
    
    # Generate original CAM
    orig_cam = model.generate_cam(image, pred_class)
    
    # Create perturbations
    perturber = ImagePerturber()
    perturbations = {
        'noise': perturber.add_gaussian_noise(image),
        'blur': perturber.add_motion_blur(image),
        'occlusion': perturber.add_occlusion(image.clone())
    }
    
    # Analyze each perturbation
    results = {}
    for name, pert_image in perturbations.items():
        with torch.no_grad():
            pert_output = model(pert_image)
        pert_class = pert_output.argmax().item()
        pert_cam = model.generate_cam(pert_image, pert_class)
        
        results[name] = {
            'image': pert_image.squeeze().permute(1,2,0).numpy(),
            'prediction': pert_class,
            'confidence': torch.softmax(pert_output, dim=1)[0][pert_class].item(),
            'cam': pert_cam
        }
        
        return orig_image, orig_cam, results
```

```python
# 4. Visualization
def visualize_results(orig_image, orig_cam, results):
    plt.figure(figsize=(20, 10))
    
    # Original image and CAM
    plt.subplot(2, 4, 1)
    plt.imshow(cv2.cvtColor(orig_image, cv2.COLOR_BGR2RGB))
    plt.title('Original Image')
    
    plt.subplot(2, 4, 2)
    plt.imshow(orig_cam, cmap='jet')
    plt.title('Original CAM')
    
    # Perturbed images and CAMs
    for i, (name, data) in enumerate(results.items(), start=3):
        plt.subplot(2, 4, i)
        plt.imshow(data['image'])
        plt.title(f"{name}\nPred: {data['prediction']}\nConf: {data['confidence']:.2f}")
        
        plt.subplot(2, 4, i+4)
        plt.imshow(data['cam'], cmap='jet')
        plt.title(f"{name} CAM")
    
    plt.tight_layout()
    plt.show()

```
